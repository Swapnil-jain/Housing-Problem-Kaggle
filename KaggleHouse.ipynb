```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
testdata=pd.read_csv("/Users/MacAir/Documents/DataScience/Datasets/home-data-for-ml-course/test.csv")
traindata=pd.read_csv("/Users/MacAir/Documents/DataScience/Datasets/home-data-for-ml-course/train.csv")
```

```python
traindata.head()
```

```python
y=traindata.SalePrice
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X_test=testdata[features]
X_main=traindata[features]
X_train, X_valid, y_train, y_valid=train_test_split(X_main,y,train_size=0.8,test_size=0.2,random_state=0)
```

```python
X_train.head()
```

```python
model_1 = RandomForestRegressor(n_estimators=50, random_state=0)
model_2 = RandomForestRegressor(n_estimators=100, random_state=0)
model_3 = RandomForestRegressor(n_estimators=100, criterion='absolute_error', random_state=0)
model_4 = RandomForestRegressor(n_estimators=200, min_samples_split=20, random_state=0)
model_5 = RandomForestRegressor(n_estimators=100, max_depth=7, random_state=0)

models=[model_1,model_2,model_3,model_4,model_5]
```

```python
def tester(model):
    model.fit(X_train,y_train)
    prediction=model.predict(X_valid)
    mae=mean_absolute_error(y_valid, prediction)
    #print (mae,type(mae))
    return mae

for i in range(0,len(models)):
    mo=float(tester(models[i]))
    print(mo)
```

```python
best_model=model_3
best_model.fit(X_main,y)
final_predictions=best_model.predict(X_test)
```

```python
output=pd.DataFrame({"Id":X_test.index,"SalePrice": final_predictions})
output.to_csv('submission_house_kaggle.csv', index=False)
```
